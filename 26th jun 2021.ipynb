Seaborn:
--------------
- Seaborn is a visualization Library in Python for statistical plotting
- It contains default attractive style
- It provides high-level interface for drawing attractive and informative statistical graphs
- Designed to work with dataframe objects in pandas

```python
import seaborn as sns
```

```python
sns.__version__
```

1. Color palette
- It is an interface to generate few colors in seaborn
    - sns.color_palette()

```python
sns.color_palette()
```

```python
sns.palplot(sns.color_palette())
```

```python
sns.palplot(sns.color_palette("deep")) # muted, colorblind
```

```python
sns.palplot(sns.color_palette("RdBu",n_colors = 8))
```

```python
sns.palplot(sns.light_palette("purple"))
```

```python
sns.palplot(sns.light_palette("purple", reverse = True))
```

```python
sns.palplot(sns.dark_palette("purple"))
```

2. Categorical Plot

```python
sns.catplot(x = "species",y = "sepal_length", data = iris)
```

```python
sns.catplot(x = "species",y = "petal_length", data = iris)
```

```python
sns.catplot(x = "species",y = "sepal_length", data = iris, kind = "box")
```

```python
sns.catplot(data = iris, kind = "box",orient = "h")
```

```python
sns.catplot(data = iris, kind = "box")
```

```python
sns.catplot(x = "species", y = "sepal_width",data = iris, kind = "violin")
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.shape
```

```python
titanic.isnull().sum()
```

```python
titanic["class"].value_counts()
```

```python
sns.catplot(x = "sex",y="survived",hue = "class",data = titanic,kind = "bar")
```

3. Joint plot 
    - combination of two plots
    - By default it have scatterplot and histogram

```python
iris.head()
```

```python
sns.jointplot(x = "petal_width",y = "sepal_width", data = iris,color = "g")
```

```python
sns.jointplot(x = "petal_width",y = "sepal_width", data = iris,kind = "kde",color = "g")
```

3. Pair Plot 
    - it displays mutiple plots at a time in a single graph
    - Pairwise relationship
    - by default it returns scatter plot

```python
sns.pairplot(iris, vars = ["sepal_length","sepal_width"])
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue = "species")
```
