Matplotlib:
-----------------
- Matplotlib is one of the library in python for constructing 2D plotting in visualization
- It is used to create high quality graphs, charts,figures etc
### Pyplot
- Pyplot is one of the most popular module in Matplotlib library
- It contains collection of functions, which is a simple and easy way to provide good user interface for constructing plots

```python
import matplotlib.pyplot as plt
```

### Types of Plots:
- Line plot
- Bar Chart
- Scatter Plot
- Box plot
- Histogram
- Piechart
- Area Plot

```python
# line plot
x = [10,12,14,16]
y = [4,5,6,7]
plt.plot(x,y)
plt.show()
```

```python
x = [10,12,14,16]
y = [4,5,6,7]
plt.plot(x,y,label = "Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Graph")
plt.legend(loc = 'upper left')
plt.show()
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,20)
x
```

```python
plt.plot(x,x**2,color = "red",linestyle = '--',label = "Line1")
plt.plot(x,x**3,color = "green",linestyle = ":",label = "Line2")
plt.plot(x,x**4,color = "yellow",linestyle="-.",label = "Line3")
plt.legend()
plt.show()
```

```python
plt.plot(x,x**2,color = "red",marker = 'o',label = "Line1")
plt.plot(x,x**3,color = "green",marker = "D",label = "Line2")
plt.legend()
plt.show()
```

```python
x1 = np.linspace(10,30,50)
plt.plot(x1,np.sin(x1-1))
plt.plot(x1,np.sin(x1-2))
plt.plot(x1,np.sin(x1-3))
plt.plot(x1,np.sin(x1-4))
plt.plot(x1,np.sin(x1-5))
plt.plot(x1,np.sin(x1-6))
plt.plot(x1,np.sin(x1-7))
plt.show()
```

```python
year = ['2019','2018','2017','2016']
gold = [1000,3000,4000,5000]
plt.bar(year,gold,color = "m")
```

```python
silver = [500,600,650,700]
plt.bar(year,gold,color = "m")
plt.bar(year,silver,color = "g")
plt.show()
```

```python
x = np.arange(5)
```

```python
# Horizontal bar graph
plt.barh(x,[90,82,57,66,75],color = ["r","g","m","pink","orange"],edgecolor = "blue")
```

```python
# Scatter plot
plt.scatter([1,3],[5,6])
```

```python
x = np.random.randint(100,1000,100)
y = np.random.randint(500,1500,100)
plt.scatter(x,y)
plt.show()
```

```python
#box plot
n = np.random.randint(1,1000,100)
plt.boxplot(n,notch = True)
```

```python
b1 = np.random.randint(1,500,50)
b2 = np.random.randint(1,700,70)
b3 = np.random.randint(1,1000,100)

plt.boxplot([b1,b2,b3],labels = ["apple","banana","mango"],patch_artist = True,
            boxprops = dict(facecolor = "pink",color = "g"),vert = False)
plt.show()
```

```python
h = [0,0.5,0.6,0.7,1,1.2,1.3,1.4,1.5,2,2.5,2.8,3,4.2,4.7,5,5.2]
bins = 4 
plt.hist(h,bins,rwidth = 0.4)
```

```python
h = [10,10.2,10.5,12,12.4,13,13.1,13.5,15,15.3,15.7,15.8]
bins = [10,11,12,13,14,15]
plt.hist(h,bins,rwidth = 0.4)
```

```python
# pie chart
daily = ["reading","sleeping","playing","watching"]
score = [40,20,10,30]
plt.pie(score,labels = daily,colors = ["c","r","m","b"])
plt.show()
```

```python
daily = ["reading","sleeping","playing","watching"]
score = [80,90,70,50]
plt.pie(score,labels = daily,autopct = "%1.2f%%",colors = ["c","r","m","b"],startangle = 90,
        explode =(0,0,0.2,0),shadow = True )
plt.show()
plt.savefig("piechart.png")
```

```python
# stack plot
days = [1,2,3,4,5]
age = [40,56,72,35,29]
weight = [95,30,70,48,69]
plt.stackplot(days,age,weight,colors = ["c","g"],
              labels = ["age","weight"])
plt.legend()
plt.show()
```
