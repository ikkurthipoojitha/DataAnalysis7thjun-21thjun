Identifying and Eliminating Outliers:
----------------------------------------------------
- Outliers are observations that are siginificantly different from other data points
- Outliers can adversely affect the training process of machine learning algorithm, resulting in a loss of accuracy.
- The mathematical formula to retrieve the outlier data
     - Interquartile Range (IQR) = Q3(quantile(0.75) - Q1(quantile(0.25))

```python
adv = pd.read_csv("https://raw.githubusercontent.com/ikkurthipoojitha/DataAnalysis7thjun-21thjun/main/Datasets/Advertising.csv")
adv.head()
```

```python
adv.plot(kind = "box")
plt.show()
```

```python
Q3 = adv.quantile(0.75)
Q1 = adv.quantile(0.25)
IQR = Q3 - Q1 
IQR
```

```python
filter_data = adv [((adv < (Q1 - 1.5 * IQR)) | (adv >(Q3 + 1.5 * IQR))).any(axis = 1)]
filter_data
```

```python
filter_data = adv [~((adv < (Q1 - 1.5 * IQR)) | (adv >(Q3 + 1.5 * IQR))).any(axis = 1)]
filter_data 
```

```python
filter_data.plot(kind = "box")
plt.show()
```
